# Advanced Regime-Aware Portfolio Research Walkthrough

This notebook walks through the end-to-end advanced research workflow in this repository.

It answers the project's core research question:

> Can changing market regimes be detected from financial time-series data, and can
> portfolio risk exposure be adjusted dynamically in response?

The notebook calls `run_advanced_research_demo_workflow` directly (the same function
backing the `run-advanced-demo` CLI command), then inspects the generated inputs, the
exported research tables, and the advanced memo it produces.


```python
# Make the repository importable and create a default outputs folder.
# Keep this cell at index 0 so later cells can import the local package.
import logging
import sys
from pathlib import Path

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger("notebook.setup")

cwd = Path.cwd().resolve()
project_root = None

# Heuristics to locate repo root: package dir, pyproject.toml, or .git
for p in (cwd,) + tuple(cwd.parents):
    if (
        (p / "regime_risk_engine").is_dir()
        or (p / "pyproject.toml").exists()
        or (p / ".git").exists()
    ):
        project_root = p
        break

if project_root is None:
    project_root = cwd

project_root_str = str(project_root)
if project_root_str not in sys.path:
    sys.path.insert(0, project_root_str)
    logger.info(f"Added project root to sys.path: {project_root_str}")

outputs_dir = project_root / "outputs" / "advanced_demo_notebook"
outputs_dir.mkdir(parents=True, exist_ok=True)
logger.info(f"Outputs directory ensured at: {outputs_dir}")

NOTEBOOK_PROJECT_ROOT = project_root
NOTEBOOK_OUTPUTS_DIR = outputs_dir

logger.info(f"Setup complete. project_root={project_root}")
```

## Run the advanced research demo workflow

`run_advanced_research_demo_workflow` synthesizes demo market data, runs regime
detection, risk analysis, and allocation comparisons, then exports a complete
research package (tables and an advanced memo) to `output_dir`.


```python
from regime_risk_engine.research.advanced_demo_workflow import (
    run_advanced_research_demo_workflow,
)

result = run_advanced_research_demo_workflow(
    output_dir=NOTEBOOK_OUTPUTS_DIR,
    analyst="Jimena Chinchilla",
)

result.export_result.memo_path
```

## Inspect generated inputs and exported tables


```python
input_result = result.input_result
export_result = result.export_result

print("Input files:")
print(f"  price data:      {input_result.price_data_path}")
print(f"  static weights:  {input_result.static_weights_path}")
print(f"  regime policy:   {input_result.regime_policy_path}")
print(f"  stress periods:  {input_result.stress_periods_path}")
print(f"  factor returns:  {input_result.factor_returns_path}")

print("\nExported research tables:")
for name, path in export_result.exported_table_paths.items():
    print(f"  {name}: {path}")
```

## Preview the advanced memo

The advanced memo is the investment-committee-style summary of the workflow's findings.


```python
memo_text = export_result.memo_path.read_text(encoding="utf-8")
print(memo_text[:2000])
```

## Interpretation guidance

- The regime intelligence profile shows how detected regimes differ in risk and return.
- The stress test summary highlights performance during historically stressful periods.
- The factor exposure table shows how allocation strategies load on common risk factors.
- The scenario terminal summary shows the distribution of simulated forward outcomes.
- Together, these outputs support the project's thesis that regime-aware allocation can
  improve risk-adjusted outcomes relative to a static benchmark.
